```python
india_data=data[data['Region']=='India']
```

```python
Confirmed=india_data['Confirmed'].sum()
Deaths=india_data['Deaths'].sum()
Recoverd=india_data['Recovered'].sum()
```

```python
labels=['Confirmed', 'Deaths', 'Recoverd']
sizes=[Confirmed, Deaths, Recoverd]
colors=['yellow', 'red', 'green']
```

```python
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Covid19 Cases in India')
plt.show()
```

```python
categories = ['Confirmed', 'Deaths', 'Recovered']
values = [33062, 1079, 8437]
```

```python
plt.figure(figsize=(8, 6))
plt.bar(categories, values, color=['Skyblue', 'Black', 'Orange'])

plt.title("India's COVID-19 Cases", fontsize=16)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Number of Cases", fontsize=12)

for i, v in enumerate(values):
    plt.text(i, v + 500, str(v), ha='center', fontsize=10)

    plt.show()
```
